```python
f = open('1_Data/analogy_words.txt')
words = f.read().split("\n")
to = open("1_Data/word_vectors.txt",'a')
n = 0
with open("/Users/dirkwulff/Dropbox (2.0)/Work/Projects/Supervision/_materials/en") as f:
    for lin in f:
        w = lin.split(" ")[0].lower()
        if len(words) == 0:
            print("Reached end")
            break
        if w in words:
            n += 1
            if n % 1000 is 0: print(n)
            to.write(lin)
            words.remove(w)
f.close()
print(n)
```
